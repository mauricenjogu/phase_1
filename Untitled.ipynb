This data contains information on movie genre,movie popularity and movie rating.

```python
#data preview
tmdb=pd.read_csv('tmdb.movies.csv',index_col=0)
tmdb.head()
```

```python
'''the analysis requires that we compare rating and popularity with movie genre which are provided in object type.
   we want to split our genre column into individual genre for easy of analysis'''
tmdb['genre_ids']=tmdb['genre_ids'].map(lambda x: x.replace('[','').replace(']',''))
tmdb['genre_ids'] =tmdb['genre_ids'].map(lambda x: x.split(','))
tmdb=tmdb.explode('genre_ids')
tmdb.head()
```

```python
#removing any spaces in our genre data
tmdb['genre_ids']=tmdb['genre_ids'].map(lambda x:x.strip())
```

```python
#changing genre_ids into names for easy indentification
genres= { '28': "Action",'12': "Adventure", '16': "Animation", '35': "Comedy",'80':"Crime",'99': "Documentary",
        '18': "Drama", '10751': "Family", '14': "Fantasy", '36': "History", '27': "Horror", '10402': "Music", '9648': "Mystery",
        '10749': "Romance", '878': "Science Fiction",'10770': "TV Movie", '53': "Thriller",'10752': "War",'37': "Western"}
tmdb['genre_ids']=tmdb['genre_ids'].map(genres)
```

```python
#getting the mean popularity per genre
tmdb_mean=tmdb.groupby('genre_ids')['popularity'].mean()
tmdb_mean
```

```python
#ploting a bar graph of the above data 
plt.figure(figsize=(8, 6))
tmdb_mean.plot(x=tmdb_mean.index,y=tmdb_mean ,kind='bar')
plt.xlabel('genre')
plt.ylabel('popularity')
plt.title('movie genre popularity');
```

```python
#getting the mean rating per genre
tmdb_mean_rating=tmdb.groupby('genre_ids')['vote_average'].mean()
tmdb_mean_rating
```

```python
#a bar graph comparing different genre and their rating
plt.figure(figsize=(8, 6))
tmdb_mean_rating.plot(x=tmdb_mean_rating.index,y=tmdb_mean_rating ,kind='bar')
plt.xlabel('genre')
plt.ylabel('rating')
plt.title('movie genre vs rating');
```
