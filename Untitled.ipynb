Analysisng Data from The Movie Database (TMDB)

This data contains information on movie genre,movie popularity and movie rating.
we are trying to see the which genre of movie has more popularity and which has better ratings in the market

```python
#data preview
tmdb=pd.read_csv('tmdb.movies.csv',index_col=0)
tmdb.head()
```

```python
'''the analysis requires that we compare rating and popularity with movie genre which are provided in object type.
   we want to split our genre column into individual genre for easy of analysis'''
tmdb['genre_ids']=tmdb['genre_ids'].map(lambda x: x.replace('[','').replace(']',''))
tmdb['genre_ids'] =tmdb['genre_ids'].map(lambda x: x.split(','))
tmdb=tmdb.explode('genre_ids')
tmdb.head()
```

```python
#removing any spaces in our genre data
tmdb['genre_ids']=tmdb['genre_ids'].map(lambda x:x.strip())
```

```python
#changing genre_ids into names for easy indentification
genres= { '28': "Action",'12': "Adventure", '16': "Animation", '35': "Comedy",'80':"Crime",'99': "Documentary",
        '18': "Drama", '10751': "Family", '14': "Fantasy", '36': "History", '27': "Horror", '10402': "Music", '9648': "Mystery",
        '10749': "Romance", '878': "Science Fiction",'10770': "TV Movie", '53': "Thriller",'10752': "War",'37': "Western"}
tmdb['genre_ids']=tmdb['genre_ids'].map(genres)
```

```python
#getting the mean popularity per genre
tmdb_mean=tmdb.groupby('genre_ids')['popularity'].mean()
tmdb_mean=tmdb_mean.sort_values(ascending=False)
tmdb_mean
```

```python
#ploting a bar graph of the above data 
plt.figure(figsize=(8, 6))
tmdb_mean.plot(x=tmdb_mean.index,y=tmdb_mean ,kind='bar')
plt.xlabel('genre')
plt.ylabel('popularity')
plt.title('TMDB movie genre popularity');
```

```python
#getting the mean rating per genre
tmdb_mean_rating=tmdb.groupby('genre_ids')['vote_average'].mean()
tmdb_mean_rating=tmdb_mean_rating.sort_values(ascending=False)
tmdb_mean_rating
```

```python
#a bar graph comparing different genre and their rating
plt.figure(figsize=(8, 6))
tmdb_mean_rating.plot(x=tmdb_mean_rating.index,y=tmdb_mean_rating ,kind='bar')
plt.xlabel('genre')
plt.ylabel('rating')
plt.title('TMDB movie genre vs rating');
```

Analysis of The Internet Movie Database(IMDB)

This data contains different tables that cointain different information on movies.In this analysis will be using
just 2 tables .One contains the information about movie names and genres while the othe cointains information on
movie rating.They can be joined by a common column which also happens to be the index of both tables.

```python
#reading the first table(title information)
imdb_movie_detail=pd.read_csv('title.basics.csv',index_col=0)
#reading the second table(rating table)
imdb_rating=pd.read_csv('title.ratings.csv',index_col=0)
#merging the 2 tables
imdb_merged=pd.merge(imdb_movie_detail,imdb_rating,on='tconst')
#preview
imdb_merged.head()
```

```python
#checking for null data in our data
imdb_merged.info()
```

The amount of null data in this table looks like a small percentage of the total data contained in the table.
i have decided to drop all rows cointaining missing data

```python
#dropping the missing values
imdb_merged.dropna(inplace=True)
```

For better analysis per movie genre .The genres column is required to contain a single genre for easy grouping
and analysis

```python
#spliting the genres column
imdb_merged['genres'] =imdb_merged['genres'].map(lambda x: x.split(','))
imdb_merged_exploded=imdb_merged.explode('genres')
#preview
imdb_merged_exploded
```

```python
#getting the average rating
imdb_average_rating=imdb_merged_exploded.groupby('genres')['averagerating'].mean()
imdb_average_rating=imdb_average_rating.sort_values(ascending=False)
imdb_average_rating
```

```python
#a bar graph comparing different genre and their rating accorging to imdb
plt.figure(figsize=(8, 6))
imdb_average_rating.plot(x=imdb_average_rating.index,y=imdb_average_rating ,kind='bar')
plt.xlabel('genre')
plt.ylabel('rating')
plt.title('IMBD movie genre vs rating');
```

checking any relationship between runtime and rating

```python
imdb_merged_exploded['runtime_minutes']=imdb_merged_exploded['runtime_minutes'].astype(int)
imdb_merged_exploded['runtime_minutes']=imdb_merged_exploded['runtime_minutes'].map(lambda x:round(x,-1))
runtime_rating=imdb_merged_exploded.groupby('runtime_minutes')['averagerating'].mean()
runtime_rating
```

```python
plt.figure(figsize=(8, 6))
runtime_rating.plot(x=runtime_rating.index,y=runtime_rating )
plt.xlim(0,1000)
plt.xlabel('runtime')
plt.ylabel('rating')
plt.title('IMBD movie duration vs rating');
```
